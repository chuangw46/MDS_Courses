```r
library(tidyverse)
library(cowplot)
library(infer)
library(taxyvr)
library(gapminder)
```

```r
ggplot(gapminder, aes(
    x = year,
    y = pop,
    color = continent)) +
    geom_line(stat = 'summary', fun = mean)
```

```r
gapminder %>% 
    filter(year==2002) %>% 
    ggplot(aes(x = lifeExp, y = pop, color = gdpPercap)) +
    geom_point() +
    facet_wrap(~continent)
```

```r
iris %>%
    select_if(is.numeric) %>%
    pivot_longer(everything()) %>%
    ggplot() + 
        aes(x = value) +
        geom_density(fill = 'SteelBlue') + 
        facet_wrap(~name, scales = 'free')
```

```r
gg_gist_multi_views <- iris %>% 
    select(-Species) %>% 
    pivot_longer(everything()) %>% 
    ggplot() + 
        aes(x = value) +
        geom_histogram() +
        facet_wrap(~name, scales = "free")

gg_gist_multi_views
```
